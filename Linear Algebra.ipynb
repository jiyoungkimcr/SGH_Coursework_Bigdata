**Question 3.** Create a 10×10 matrix of zeros and then "frame" it with a border of ones.

```
x = np.zeros((10,10))
print (x)
```

```
# Put '1' in first & last columns
x[:10, ::9] = 1  # Put 1 in all 10 rows by every other 9 column (first&last column)
x[::9, :10] = 1  # Put 1 in first&last row in every row
print(x)
```

**Question 9.** Write a function which creates an n×n matrix with (i,j)-entry equal to i+j.

```
# Method 1 - Use 'fromfunction'
# (*caution: in NumPy, row index(i) and column index(j) start from '0' not '1')
n=3   # the below result is when 'n=3' (of course, 'n' can be any positive integer)
np.fromfunction(lambda i, j: i+j, (n,n))
```

```
# Method 2 - Create empty nxn array and fill it using 'for' loop
n=3 
X = np.empty(shape=(n,n), dtype=int)

for i in range(n):
  for j in range(n):
    X[i,j] = i+j
    
X
```

**Question 10.** Write a function which creates an n×n matrix with rows having subsequent values multiplied by the row's number. For example for n = 4:

[[0, 1, 2, 3], [0, 2, 4, 6], [0, 3, 6, 9], [0, 4, 8, 12]]

```
n=4 #the result below is the result of the function when n=4
np.fromfunction(lambda i, j: [i-i,j]*(i+1), (n,n)) 
```
